### Prepare data

```python
# Drop unnecessary columns
df.drop(['Name'], axis=1, inplace=True)
df.head()
```

### Ordinal encoder

```python
# List unique values for Education
df['Education'].unique()
```

```python
# Encode Education column
oe = OrdinalEncoder(categories=[['H.S.', 'B.S.', 'M.S.','Ph.D']])     # Arrange values according to relative value
df['Education'] = oe.fit_transform(df[['Education']])
df.head()
```

### Dummy (one hot) encoding
For simplicity, we'll use pd.get_dummies() instead of OneHotEncoder()

#### Step-by-step dummy (one hot) encoding

```python
# List unique values for Gender
df['Gender'].unique()
```

```python
df.head()
```

```python
# Generate encodings for Gender
df_ohe1 = pd.get_dummies(df['Gender'], dtype=int)
df_ohe1.head()
```

```python
# Join dataframe with Gender-encoded dataframe
df2 = df.join(df_ohe1)
df2.head()
```

```python
# Drop Gender column since it is now encoded
df2.drop(['Gender'], axis=1, inplace=True)
df2.head()
```

#### Simpler dummy (one hot) encoding syntax

```python
# List unique values for Marital Status
df2['Marital_Status'].unique()
```

```python
# Clearer way to perform one hot encoding
df2 = pd.get_dummies(df2, columns=['Marital_Status'], dtype=int)
df2.head()
```

```python
# Drop column State
df2.drop(['State'], axis=1, inplace=True)
df2.head()
```

### Label encoding

```python
# Encode Employment Status column
le = LabelEncoder()
df2['Employment_Status']= le.fit_transform(df2['Employment_Status'])
df2.head()
```
