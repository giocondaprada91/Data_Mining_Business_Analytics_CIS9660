### Import libraries

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

### Load data

```python
# Load data file (iPhone_purchases.csv) into dataframe
df = pd.read_csv('https://raw.githubusercontent.com/vjavaly/Baruch-CIS-9660/main/data/iPhone_purchases.csv')
```

### Examine data

```python
df.shape
```

```python
df.head()
```

#### Dummy (one-hot) encode column Gender

```python
df = pd.get_dummies(df, columns=['Gender'])
df.head()
```

### Separate independent and dependent variables

```python
X = df.drop('Purchased_iPhone', axis = 1)
X.head()
```

```python
y = df['Purchased_iPhone']
y.head()
```

### Normalize independent variables

```python
min_max_scaler = MinMaxScaler()
X_scaled = min_max_scaler.fit_transform(X)
X_scaled[:5]
```

### Split data into training and test sets

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, stratify=y,
                                                    test_size=0.2, random_state=42)
```

### Train model (with default hyperparameters)

```python
knn = KNeighborsClassifier()
knn.get_params()
```

```python
knn.fit(X_train, y_train)
```

### Calculate accuracy score (with default hyperparameters)

```python
# Predict using the test set
y_pred = knn.predict(X_test)
```

```python
# Print model accuracy score
accuracy_score_1 = accuracy_score(y_test, y_pred)
print(f"Accuracy = {round((accuracy_score_1 * 100), 4)}%")
```

```python
# Print model accuracy score (alternate method)
accuracy_score_2 = knn.score(X_test, y_test)
print(f"Accuracy = {round((accuracy_score_2 * 100), 4)}%")
```

### Train model with different hyperparameters (experiment 1)

```python
knn = KNeighborsClassifier(n_neighbors=3, weights='distance', metric='manhattan')
knn.fit(X_train, y_train)
```

### Calculate accuracy score (experiment 1)

```python
# Predict using the test set
y_pred = knn.predict(X_test)

# Print model accuracy score
accuracy_score_1 = accuracy_score(y_test, y_pred)
print(f"Accuracy = {round((accuracy_score_1 * 100), 4)}%")
```

### Train model with different hyperparameters (experiment 2)

```python
knn = KNeighborsClassifier(n_neighbors=7, weights='distance', metric='euclidean')
knn.fit(X_train, y_train)
```

### Calculate accuracy score (experiment 2)

```python
# Predict using the test set
y_pred = knn.predict(X_test)

# Print model accuracy score
accuracy_score_1 = accuracy_score(y_test, y_pred)
print(f"Accuracy = {round((accuracy_score_1 * 100), 4)}%")
```
