### Import libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

### Load data

```python
# Read data file (advertising.csv) into a dataframe
df = pd.read_csv('https://raw.githubusercontent.com/vjavaly/Baruch-CIS-9660/main/data/advertising.csv')
```

### Examine data

```python
# Review dataframe shape
df.shape
```

```python
# Review first 10 rows
df.head(10)
```

### Prepare data for model training

```python
# Drop non-numeric columns: City, State, Region
df.drop(['City','State', 'Region'], axis=1, inplace= True)
```

```python
# Display first few rows of updated dataframe
df.head()
```

### Separate independent and dependent variables

```python
# Independent variables: All columns EXCEPT Sales
# Dependent variable: Sales
X= df.drop('Sales', axis=1)
y= df['Sales']
```

```python
# Display first few rows of independent variables
X.head()
```

```python
# Display first few rows of dependent variable
y.head()
```

### Split data into training and test sets

```python
# 70% training set, 30% test set
random_state = 42
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,
                                                    random_state=random_state)
```

### Train Linear Regression model

```python
model = LinearRegression()
model.get_params()
model.fit(X_train, y_train)
```

### Calculate model performance for the test set

```python
# Make predictions against the test set
y_test_predict = model.predict(X_test)
mse = mean_squared_error(y_test, y_test_predict)
rmse = np.sqrt(mse)
```

```python
# Display MSE and RMSE
print("Model performance for test set")
print("----------------------------------")
print("MSE is {}".format(round(mse,2)))
print("RMSE is {}".format(round(rmse,2)))
```
