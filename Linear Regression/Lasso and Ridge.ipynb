### Initialize global variables

```python
global RANDOM_STATE, ALPHA
RANDOM_STATE = 42
ALPHA = 10
```

### Import libraries

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', '{:.4f}'.format)
```

### Load data

```python
# Read data file (car_prices_1.csv) into a dataframe
df = pd.read_csv('https://raw.githubusercontent.com/vjavaly/Baruch-CIS-9660/main/data/car_prices.csv',
                 index_col='car_ID')
```

### Examine data

```python
df.shape
```

```python
df.head()
```

### Separate independent and dependent variables

```python
# Independent variables: All columns except 'price'
X = df.drop('price', axis = 1)

# Dependent variable: price
y = df['price']
```

### Split data into training and test sets

```python
# 70% training set, 30% test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,
                                                    random_state=RANDOM_STATE)
```

### Train Linear Regression model

```python
# Instantiate linear regression model
model_lin = LinearRegression()

# Print model hyperparameters
model_lin.get_params()
```

```python
# Fit linear regression model
model_lin.fit(X_train, y_train)
```

```python
# Save linear regression model parameters (coefficients)
lin_coefs = pd.Series(model_lin.coef_, index=df.columns[0:37]).to_frame()
lin_coefs.rename(columns={lin_coefs.columns[0]: 'linear regression coeffs'}, inplace=True)
```

**model_lin.coef_:** This comes from a linear regression model (model_lin). Once a linear regression model is trained, coef_ contains the coefficients of the features used in the model. These coefficients represent the strength and direction of the relationship between the features (independent variables) and the target variable (dependent variable).

pd.Series() is used to create a Pandas Series object.
The first argument model_lin.coef_ is a list or array of the linear regression coefficients.

index=df.columns[0:37]: This specifies that the index (or labels) for the Series should be the first 37 column names from the DataFrame df. So, the coefficients will be matched with the names of the first 37 features/columns in the DataFrame.

.to_frame(): This converts the Series into a DataFrame. A DataFrame has a 2D table structure (rows and columns), whereas a Series is a 1D object.
In this case, it takes the coefficients and puts them into a DataFrame with one column.


### Train Lasso Regression model

```python
# Instantiate lasso regression model
model_lasso = Lasso(alpha=ALPHA, random_state=RANDOM_STATE)

# Print model hyperparameters
model_lasso.get_params()
```

```python
# Fit lasso regression model
model_lasso.fit(X_train, y_train)
```

```python
# Save lasso regression model parameters (coefficients)
lasso_coefs = pd.Series(model_lasso.coef_, index=df.columns[0:37]).to_frame()
lasso_coefs.rename(columns={lasso_coefs.columns[0]: 'lasso regression coeffs'}, inplace=True)
```

### Train Ridge Regression model

```python
# Instantiate ridge regression model
model_ridge = Ridge(alpha=ALPHA, random_state=RANDOM_STATE)

# Print model hyperparameters
model_ridge.get_params()
```

```python
# Fit ridge regression model
model_ridge.fit(X_train, y_train)
```

```python
# Save linear regression model parameters (coefficients)
ridge_coefs = pd.Series(model_ridge.coef_, index=df.columns[0:37]).to_frame()
ridge_coefs.rename(columns={ridge_coefs.columns[0]: 'ridge regression coeffs'}, inplace=True)
```

```python
all_coefs = pd.concat([lin_coefs, lasso_coefs, ridge_coefs], axis=1)
all_coefs
```
