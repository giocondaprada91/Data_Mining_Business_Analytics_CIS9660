### Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
```

### Load data

```python
df = pd.read_excel('https://raw.githubusercontent.com/vjavaly/Baruch-CIS-9660/main/data/CoffeeShopSales.xlsx')
df.shape
```

### Examine data

```python
df.sample(5)
```

### Prepare data
Convert data for market basket analysis

```python
# Create pivot table
#  The purpose of creating the pivot table is to transform the original
#   transactional dataset into a format suitable for market basket analysis
df_pivot = df.pivot_table(index='transaction_number',
                          columns ='item',
                          values = 'amount',
                          aggfunc='sum').fillna(0)
df_pivot.shape
```

```python
pd.set_option('display.max_columns', None)
```

```python
df_pivot.sample(5)
```

```python
# Change the data type from float to integer
df_pivot = df_pivot.astype(int)
df_pivot.head()
```

```python
# Change the value if it is > 1 into 1, and if it is < 1 into 0
def encode(x):
    if x <=0:
        return 0
    else:
        return 1

df_pivot = df_pivot.applymap(encode)
```

### Find frequently occurring itemsets using Apriori algorithm
Support is the minimum support threshold.  
It sets the minimum occurrence frequency (as a proportion of the total transactions) for an item set to be considered "frequent".

```python
# Set support threshold to 0.01
support = 0.01

frequent_items = apriori(df_pivot,
                         min_support=support,
                         use_colnames=True)
frequent_items.sort_values('support', ascending=False)
```

# Define association rules

```python
# Set lift threshold to 1
metric = 'lift'
lift = 1

rules = association_rules(frequent_items, num_itemsets=None,
                          metric=metric,
                          min_threshold=lift)
rules.reset_index(drop=True).sort_values('confidence', ascending=False, inplace = True)
rules
```

From the association rules above we can see that when a customer buys a almond croissant, they also tend to buy cappuccino iced

Lets say that in this coffee shop, the sales of almond croissants are not meeting the sales target. By using the market basket analysis, we can try to create a promotion such as buy one cappuccino and get a 25% discount on almond croissants.

In essence, market basket analysis provides valuable insights that businesses can leverage to enhance sales, customer experience, and operational efficiency by understanding customer purchasing behaviors and preferences.
